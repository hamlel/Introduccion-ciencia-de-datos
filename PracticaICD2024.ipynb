## El conjunto de datos

El conjunto de datos fue recopilado a partir de las erupciones del géiser Old Faithful, ubicado en el parque nacional de Yellowstone, Wyoming, EE.UU. Se consideran dos variables: el tiempo de la erupción y el tiempo entre erupciones, ambos en minutos. El conjunto de datos fue presentado originalmente en Hazzalini y Bowman (1990) y Hardle (1991).

Para esta práctica, se cuenta con 272 observaciones y dos variables etiquetadas como "Time eruption" y "Time Waiting". Ambas variables son de tipo cuantitativo-continuo.

```python
import pandas as pd
from matplotlib import pyplot as plt
import numpy as np
import seaborn as sns
# en este apartado se cargan las librerías que necesitaremos para
#realizar el análisis exploratorio de los datos.
```

```python
url="https://raw.githubusercontent.com/gchoi/Dataset/master/OldFaithful.csv"
# en este apartado se guarda la dirección del conjunto de datos de Github
#y se procede a leer el conjunto de datos, los primeros ocho registros
df = pd.read_csv(url)
df.head(8)
```

```python
df.shape
# En este apartado identificamos si se presentan valores perdidos
# Así también se identifican el número de registros, columnas y el tipo de datos.
```

De acuerdo con las medidas de tendencia central, se identifica que el tiempo promedio de las erupciones es de 3.48 minutos, mientras que el tiempo promedio entre erupciones es de 70.89 minutos. Es importante destacar que la dispersión del tiempo de las erupciones es menor (1.14 minutos) en comparación con la dispersión del tiempo entre erupciones (13.59 minutos). Además, el 75% de los tiempos de erupción fueron menores a 4.45 minutos. A su vez, el 75% de los tiempos entre erupciones fueron menores a 82 minutos, lo que representa una diferencia significativa en la dispersión de las variables.

```python
df.describe(include = "all")
#Posteriormente se calculan las medidas estadísticas de centralidad y dispersión de las dos variables
```

```python
sns.violinplot(data=df,fill=False);
#Posteriormente se gráfican los violinplots que pueden ser enriquecidos
#Con la tabla de medidas de centralidad y dispersión ejecutadas anteriormente.
```

Continuando con los resultados exploratorios, se observa que las variables no siguen un comportamiento cercano a una distribución normal. Además, se identifican dos agrupamientos bien definidos en la relación entre el tiempo entre erupciones y el tiempo de las erupciones.

```python
sns.pairplot(df,plot_kws=dict(marker="+", linewidth=1),
    diag_kws=dict(fill=False),);
# En este apartado se realiza un gráfico relacionando
#las dos variables del conjunto de datos
```

Por otro lado, al examinar las curvas de densidad kernel, se observa claramente una marcada diferencia entre el tiempo de las erupciones y el tiempo entre las erupciones. El tiempo de las erupciones varía de 1 a 5 minutos, mientras que el tiempo entre erupciones oscila entre 40 y 100 minutos. Además, se utilizan diferentes anchos de banda del kernel para ajustar el comportamiento de los datos: 0.01, 0.1 y 0.9.

```python
#sns.kdeplot(data=df,fill=True)
#Con este comando se calcula la densidad
sns.kdeplot(data=df,bw_adjust=0.01)
#De la distribución de las variables de tiempo de la erupcion y tiempo entre erupciones.
```

```python
sns.kdeplot(data=df,bw_adjust=0.1)
```

```python
sns.kdeplot(data=df,bw_adjust=0.9)
```

Por último, se presenta un ajuste de densidad kernel bivariado entre el tiempo de las erupciones y el tiempo entre erupciones, donde se identifica un interesante agrupamiento en los comportamientos de las variables de estudio. Estos resultados son útiles para mejorar las estimaciones de riesgo volcánico.

```python
sns.kdeplot(x='TimeEruption', y='TimeWaiting', data=df,fill=True);
#Por último se ejecuta un estimación de densidad kernel para identificar las agrupaciónes entre los tiempos de erupciones y tiempo entre erupciones del conjunto de datos analizado.
```
